```python
from pmdarima.arima import auto_arima
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
shamp = pd.read_csv("C:/Users/Administrator.DAI-PC2/Desktop/ML/Day14/sales-of-shampoo-over-a-three-ye.csv")
shamp
shamp.plot()
plt.show()
```

```python
y = shamp['Sales_of_shampoo']
y_train = y[:-6]
y_test = y[-6:]
```

```python
model = auto_arima(y_train, trace=True, error_action='ignore',  suppress_warnings=True)

forecast = model.predict(n_periods=len(y_test))
forecast = pd.DataFrame(forecast,index = y_test.index,columns=['Prediction'])
forecast
```

```python
# plot results
plt.plot(y_test, label='Test Data')
plt.plot(forecast, color='red', label="Forecast")
plt.legend(loc='best')
plt.show()
```

```python
#plot the predictions for validation set
plt.plot(y_train, label='Train',color="blue")
plt.plot(y_test, label='Valid',color="pink")
plt.plot(forecast, label='Prediction',color="purple")
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(y_test, forecast))
print('Test RMSE: %.3f' % rms)
```

```python
### SARMIA
model = auto_arima(y_train, trace=True, error_action='ignore', suppress_warnings=True,seasonal=True,m=12)
forecast = model.predict(n_periods=len(y_test))
forecast = pd.DataFrame(forecast,index = y_test.index, columns=['Prediction'])


# plot results
plt.plot(y_test, label='Test Data')
plt.plot(forecast, color='red', label="Forecast")
plt.legend(loc='best')
plt.show()
```

```python
#plot the predictions for validation set
plt.plot(y_train, label='Train',color="blue")
plt.plot(y_test, label='Valid',color="pink")
plt.plot(forecast, label='Prediction',color="purple")
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(y_test, forecast))
print('Test RMSE: %.3f' % rms)
```
