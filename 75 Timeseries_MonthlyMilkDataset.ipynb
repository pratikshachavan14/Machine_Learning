```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Administrator.DAI-PC2/Desktop/ML/Day13/monthly-milk-production-pounds-p.csv", index_col=0)
df
df.index = pd.to_datetime(df.index).to_period("M")
```

```python
df.plot()
plt.title("Monthly Milk Production")
plt.xlabel("Months")
plt.show()
```

```python
downsampled = df.resample('Q').sum()
downsampled.index.rename('Quarter', inplace=True)
downsampled.plot()
plt.title("Quarterly Milk Production")
plt.xlabel("Quarters")
plt.show()
```
