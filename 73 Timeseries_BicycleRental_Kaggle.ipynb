```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Administrator.DAI-PC2/Desktop/ML/Day13/train.csv",  parse_dates=['datetime'])
df
#df.index = pd.to_datetime(df.index).to_period("D")   in df index_col=0
df.set_index('datetime', inplace=True)
df
```

```python
casual = df['casual']
mon_cas = casual.resample('M').sum()
mon_cas.plot()
plt.title("Monthly casual rentals")
plt.xlabel("Months")
plt.show()
```

```python
mon_cas = casual.resample('Q').sum()
mon_cas.index.rename('Quarter', inplace=True)
mon_cas.plot()
plt.title("Quarterly casual rentals")
plt.xlabel("Quarter")
plt.show()
```

```python
registered = df['registered']
mon_reg = registered.resample('Q').sum()
mon_reg.index.rename('Quarter', inplace=True)
mon_reg.plot()
plt.title("Quarterly registered rentals")
plt.xlabel("Quarter")
plt.show()
```

```python
registered = df['registered']
mon_reg = registered.resample('M').sum()
mon_reg.index.rename('Monthly', inplace=True)
mon_reg.plot()
plt.title("Monthly registered rentals")
plt.xlabel("Monthly")
plt.show()
```
