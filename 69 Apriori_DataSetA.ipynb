```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = []
with open(r"C:\Users\Administrator.DAI-PC2\Desktop\ML\Day12\DataSetA.csv", "r") as f:df=f.read()
df = df.split("\n")
```

```python
df_list = []
for i in df:
    df_list.append(i.split(","))
```

```python
te = TransactionEncoder()
te_ary = te.fit(df_list).transform(df_list)
te_ary
```

```python
fp_df = pd.DataFrame(te_ary, columns = te.columns_)
fp_df
```

```python
fp_df = pd.DataFrame(te_ary[:,1:], columns = te.columns_[1:])
fp_df
```

```python
#create frequency itemsets
itemsets = apriori(fp_df, min_support=0.01, use_colnames=True)
```

```python
#and convert into rules
rules = association_rules(itemsets, metric='confidence', min_threshold=0.5)
rules.sort_values(by=['lift'], ascending=False).head(6)
rules_df = rules[rules['lift']>1]

rules_df[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```
