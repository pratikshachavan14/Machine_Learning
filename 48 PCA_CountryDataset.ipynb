```python
df = pd.read_csv("C:/Users/Administrator.DAI-PC2/Desktop/ML/Day10/Country-data.csv", index_col=0)


pca = PCA().set_output(transform='pandas')
std_scaler = StandardScaler().set_output(transform='pandas')
m_scaled = std_scaler.fit_transform(df)
components = pca.fit_transform(m_scaled)
```

```python
from pca import pca
model = pca()
results = model.fit_transform(m_scaled,
                             col_labels = df.columns,
                             row_labels = list(df.index))
```

```python
model.biplot(label=True, legend=True)
for i in np.arange(0, df.shape[0]):
    plt.text(components.values[i,0],
            components.values[i,1],
            list(df.index)[i])
plt.show()
```
