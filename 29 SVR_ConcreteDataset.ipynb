```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, accuracy_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("C:/Users/Administrator.DAI-PC2/Desktop/ML/Day 5/Concrete_Data.csv")

y = df['Strength']
X = df.drop('Strength', axis = 1)
```

```python
svr = SVR(C= 0.1, kernel = 'linear', epsilon = 0.1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=24)

kfold = KFold(n_splits = 5, shuffle = True, random_state=24)

svr.fit(X_train, y_train)
```

```python
#linear, do not run, doesn't end....

std_scaler = StandardScaler()
std_mm = MinMaxScaler()
pipe = Pipeline([('SCL', None), ('SVR',svr)])
params = {'SVR__C': np.linspace(0.001, 5, 3), 'SCL':[std_scaler, std_mm, None], 'SVR__epsilon': np.linspace(0.001, 5, 3)}
gcv = GridSearchCV(pipe, param_grid = params, cv = kfold, scoring = 'r2', verbose = 2)
#gcv.fit(X, y)
#print(gcv.best_score_)
#print(gcv.best_params_)
```

```python
#kernel = poly

svr1 = SVR(C= 0.1, kernel = 'poly')

pipe = Pipeline([('SCL', None), ('SVR',svr1)])
params = {'SVR__C': np.linspace(0.001, 5, 3), 'SCL':[std_scaler, std_mm, None], 'SVR__degree' : [2,3], 'SVR__coef0': np.linspace(0, 3, 5),  'SVR__epsilon': np.linspace(0.001, 5, 3)}
gcv_poly = GridSearchCV(pipe, param_grid = params, cv = kfold, scoring = 'r2', verbose = 2)
gcv_poly.fit(X, y)
print(gcv_poly.best_score_)
print(gcv_poly.best_params_)
```

```python
#kernel = rbf

svr2 = SVR(C= 0.1, kernel = 'rbf')

pipe = Pipeline([('SCL', None), ('SVR',svr2)])
params = {'SVR__C': np.linspace(0.001, 5, 3), 'SCL':[std_scaler, std_mm, None], 'SVR__degree' : [2,3], 'SVR__gamma': np.linspace(0.001, 5, 3), 'SVR__epsilon': np.linspace(0.001, 5, 3)}
gcv_rbf = GridSearchCV(pipe, param_grid = params, cv = kfold, scoring = 'r2', verbose = 2)
gcv_rbf.fit(X, y)
print(gcv_rbf.best_score_)
print(gcv_rbf.best_params_)
```
